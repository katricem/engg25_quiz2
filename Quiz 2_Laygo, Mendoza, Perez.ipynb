# Question 1

Find the solution of the initial value problem: $ 2y'' - 3y' + y = 0 $, $y(0) = 2$, $y'(0) = \frac{1}{2}$. Then determine the maximum value of the solution and also find the point where the solution is zero.

## Solution:
$ 2 r^2 e^{rt} - 3 re^{rt} + e^{rt} = 0 $

$ e^{rt} (2r^2 - 3r +1) = 0 $

Characteristic equation: $2r^2 - 3r +1 = 0$

$ r = \frac{-(-3) \pm \sqrt{(-3)^2 - 8}}{4}$

$r_1 = 1, r_2 = \frac{1}{2}$


$ y(t) = c_1 e^{1t} + c_2 e^{\frac{1}{2}t} $

$ y(0) = 2 $

$ y(0) = c_1 e^0 + c_2 e^0 $

$ c_1 + c_2 = 2 $ (eqn. 1)

$ y'(t) = c_1 e^t + c_2 \frac{1}{2} e^{\frac{1}{2}t} $

$ y'(0) = \frac{1}{2} $

$ y'(0) = c_1 e^0 + c_2 \frac{1}{2} e^{\frac{1}{2}0} $

$ c_1 + c_2 \frac{1}{2} = \frac{1}{2} $ (eqn. 2)

Inputting equation 1 and 2 into the calculator we get: $ c_1 = -1 $, $ c_2 = 3 $

Solution: $ y(t) = -e^t + 3e^{\frac{1}{2}t} $

### finding the maximum value of the solution:

$ y'(t) = -e^t + \frac{3}{2} e^{\frac{t}{2}} $

$ y'(t) = 0$ to get maximum value

$ \frac{3}{2} e^{\frac{t}{2}} - e^t = 0 $

$ \frac{3}{2} e^{\frac{t}{2}} = e^t $

$ 3 e^{\frac{t}{2}} = 2 e^t $

$ e^{\frac{1}{2}t} = \frac{3}{2} $

$ \frac{1}{2}t = ln(\frac{3}{2}) $

$ t = 2ln(\frac{3}{2}) \approx 0.81$

$ y(0.81) = -e^{0.81} + 3e^{\frac{1}{2}(0.81)} $

the maximum value is $ y(0.81) = 2.249999513 $ or approx. $ 2.25 $

### where the solution is 0

$ y(t) = 0 $

$ 3e^{\frac{1}{2}t} - e^t = 0 $

$ 3e^{\frac{1}{2}t} = e^t $

$ 3 = e^{\frac{1}{2}t} $

$ \frac{1}{2}t = ln 3$

the solution is zero when $ t = 2ln3 $


# Question 2
Determine the longest interval in which the given initial value problem is certain to have a unique twice-differentiable solution: $ y'' + (cos t)y' + 3(ln|t|)y = 0 $, $ y(2) = 3 $, $ y'(2) = 1$. Do not try to find the solution.

## solution:

$ p(t) = cos t $

$ q(t) = ln |t| $

$p(t)$ is continuous for all real numbers, while $q(t)$ is continuous over $(-\infty, 0)$ and $(0, \infty)$.

The initial values given show a value of 2, therefore the longest interval is $(0, \infty)$.
